```python
import numpy as np
arr = np.arange(1, 21)
print("Array:")
print(arr)	

# 2. Find the shape of the array
print(arr.shape)
# 3. Find the maximum and minimum values
print(arr.max())
print(arr.min())

# 4. Calculate the mean and standard deviation
print(arr.mean())
print(arr.std())

# 5. Reshape the array into a 4 x 5 matrix
print("Reshaped Matrix:")
matrix = arr.reshape(4, 5)
print(matrix)
# 6. Display the second row
print("Second Row:")
print(matrix[1])
# 7. Display the third column
print("Third Column:")
print(matrix[:, 2])
```

```python
import numpy as np
A = np.array([10, 20, 30, 40, 50])
B = np.array([5, 10, 15, 20, 25])
print(A + B)
print(A - B)
print(A * B)
print(A / B)
print("\nSum of elements in A:")
print(A.sum())
```

```python
import pandas as pd
data = {
    "Roll No": [101, 102, 103, 104, 105],
    "Name": ["Ravi", "Priya", "Kiran", "Anu", "Sita"],
    "Marks": [85, 92, 78, 88, 95]
}
df = pd.DataFrame(data)
print("Student Data:")
print(df)

print("\nFirst Three Records:")
print(df.head(3))

print("\nAverage Marks:")
print(df["Marks"].mean())

print("\nStudent with Highest Marks:")
highest = df["Marks"].max()
print(df[df["Marks"] == highest])

print("\nStudent with Lowest Marks:")
print(df[df["Marks"] == df["Marks"].min()])

grade = []
for marks in df["Marks"]:
    if marks >= 90:
        grade.append("A")
    elif marks >= 80:
        grade.append("B")
    else:
        grade.append("C")
df["Grade"] = grade
print("\nDataFrame with Grade:")
print(df)
```
