```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv(r"C:\Users\Harsha vardhini\Downloads\ML_Lab-2.csv")
#first 5 records
display(df.head(5))
#last 5 records
display(df.tail(5))
#display Rows and Columns
print(df.shape)
#column names
print(df.columns.tolist())
#dataset information using info()
display(df.info())
```

```python
#mean
avg_attendance=df['Attendance'].mean()
print(f"Avg Attendance is {avg_attendance}") 
#internal
avg_internal = df['InternalMarks'].mean()
print(f"Avg Internal Marks is {avg_internal}")
avg_external = df['ExternalMarks'].mean()
print(f"Avg External Marks is {avg_external}")
max_internal = df['InternalMarks'].max()
print(f"Maximum internal Marks is {max_internal}")
min_internal = df['InternalMarks'].min()
print(f"Minimum internal Marks is {min_internal}")

max_external = df['ExternalMarks'].max()
print(f"Maximum external Marks is {max_external}")

min_external = df['ExternalMarks'].min()
print(f"Minimum external Marks is {min_external}")

std_marks = np.std(df['InternalMarks'],ddof=1)
print(f"Standard Deviation of Internal Marks{std_marks}")
df.describe()
```

```python
#data manipulation
df['TotalMarks'] = df['InternalMarks'] + df['ExternalMarks']
display(df) 
df['percentage'] = (df['TotalMarks']/100)*100
display(df)
```

```python
#filtering ans search
high_attendance_students = df[df['Attendance']>90]
display(high_attendance_students)

high_totalmarks_students = df[df['TotalMarks']>90]
display(high_totalmarks_students)
female_students = df[df['Gender'] == 'F']
display(female_students)

male_students = df[df['Gender'] == 'M']
display(male_students)

external_marks_less_50 = df[df['ExternalMarks'] < 50]
display(external_marks_less_50 )
```

```python
#ranking and comparision
highest_score = df[df['TotalMarks'] ==df['TotalMarks'].max()]
display(highest_score)

lowest_score = df[df['TotalMarks'] == df['TotalMarks'].min() ]
display(lowest_score)
#Botom three performance students
bottom_three = df.nsmallest(3,'TotalMarks')
display(bottom_three)
#Top three performance students
top_three = df.nlargest(3,'TotalMarks')
display(top_three)
#students above class average
class_average = df['TotalMarks'].mean()
above_class_avg = df[df['TotalMarks'] > class_average]
display(above_class_avg)
```

```python
# Avg total marks of male students
avg_male_marks = df[df['Gender'] == 'M']['TotalMarks'].mean()
print(f"Average marks in male is {avg_male_marks}")

#Avg total Marks of female students
avg_female_marks = df[df['Gender'] == 'F']['TotalMarks'].mean()
print(f"Average marks in female is {avg_female_marks}")

#Attendance of male and female
attendance_comparision = df.groupby('Gender')['Attendance'].mean()
print(f"Average attendance by gender:\n{attendance_comparision}")
print("Girls have better attendance compared to boys")
import matplotlib.pyplot as plt
```

```python
#plot a bargraph for total marks
plt.bar(df['Name'],df['TotalMarks'])
plt.xlabel('Students')
plt.ylabel('Total Marks')
plt.title('Total Marks of Students')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
#histogram for attendance
plt.hist(df['Attendance'])
plt.xlabel('Attendance')
plt.ylabel('Frequency')
plt.title('Histogram of Attendance')
plt.show()
#pie chart for gender distribution
gender_counts = df['Gender'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index)
plt.title('Gender Distribution')
plt.show()
#scatter plot between attendance and total marks
plt.scatter(df['Attendance'], df['TotalMarks'])
plt.xlabel('Attendance')
plt.ylabel('Total Marks')
plt.title('Attendance vs Total Marks')
plt.show()
#Interpretting relation between Attendance and Total Marks
correlation = df['Attendance'].corr(df['TotalMarks'])
print(f"Correlation between Attendance and Total Marks: {correlation}")
```
